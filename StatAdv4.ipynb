# Differentiate between one-tailed and two-tailed tests.

```python
# This is because a two-tailed test uses both the positive and negative tails of the distribution. In other words, it tests for
# the possibility of positive or negative differences. A one-tailed test is appropriate if you only want to determine if there
# is a difference between groups in a specific direction.

# One Tail Test:
# A test of any statistical hypothesis, where the alternative hypothesis is one-tailed either right-tailed or left-tailed.
# For one-tailed, we use either > or < sign for the alternative hypothesis.
# When the alternative hypothesis specifies a direction then we use a one-tailed test.
# Critical region lies entirely on either the right side or left side of the sampling distribution.
# Here, the Entire level of significance (α) i.e. 5% has either in the left tail or right tail.
# Rejection region is either from the left side or right side of the sampling distribution.
# It checks the relation between the variable in a singles direction. 
# It is used to check whether the one mean is different from another mean or not.

# Two tail Test:
# A test of a statistical hypothesis, where the alternative hypothesis is two-tailed.
# For two-tailed, we use ≠ sign for the alternative hypothesis.
# If no direction is given then we will use a two-tailed test.
# Critical region is given by the portion of the area lying in both the tails of the probability curve of the test statistic.
# It splits the level of significance (α) into half.
# Rejection region is from both sides i.e. left and right of the sampling distribution.
# It checks the relation between the variables in any direction.
# It is used to check whether the two mean different from one another or not.
```

# Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error


```python
# A Type I error happens when you get false positive results: you conclude that the drug intervention improved symptoms when
# it actually didn’t. These improvements could have arisen from other random factors or measurement errors.

# A Type II error happens when you get false negative results: you conclude that the drug intervention didn’t improve symptoms 
# when it actually did. Your study may have missed key indicators of improvements or attributed any improvements to other
# factors instead.


# Type I Error:
    
# A Type I error means rejecting the null hypothesis when it’s actually true. It means concluding that results are statistically
# significant when, in reality, they came about purely by chance or because of unrelated factors.

# The risk of committing this error is the significance level (alpha or α) you choose. That’s a value that you set at the
# beginning of your study to assess the statistical probability of obtaining your results (p value).

# The significance level is usually set at 0.05 or 5%. This means that your results only have a 5% chance of occurring, or less,
# if the null hypothesis is actually true.

# If the p value of your test is lower than the significance level, it means your results are statistically significant and 
# consistent with the alternative hypothesis. If your p value is higher than the significance level, then your results are
# considered statistically non-significant.
# Ex.

# In our clinical study, you compare the symptoms of patients who received the new drug intervention or a control treatment.
# Using a t test, you obtain a p value of .035. This p value is lower than your alpha of .05, so you consider your results
# statistically significant and reject the null hypothesis.
# However, the p value means that there is a 3.5% chance of your results occurring if the null hypothesis is true. Therefore,
# there is still a risk of making a Type I error.

# Type 2 Error:

# A Type II error means not rejecting the null hypothesis when it’s actually false. This is not quite the same as “accepting” the
# null hypothesis, because hypothesis testing can only tell you whether to reject the null hypothesis.

# Instead, a Type II error means failing to conclude there was an effect when there actually was. In reality, your study may 
# not have had enough statistical power to detect an effect of a certain size.

# Power is the extent to which a test can correctly detect a real effect when there is one. A power level of 80% or higher is
# usually considered acceptable.

# The risk of a Type II error is inversely related to the statistical power of a study. The higher the statistical power, the
# lower the probability of making a Type II error.

# Ex.

# When preparing your clinical study, you complete a power analysis and determine that with your sample size, you have an 80%
# chance of detecting an effect size of 20% or greater. An effect size of 20% means that the drug intervention reduces symptoms
# by 20% more than the control treatment.
# However, a Type II may occur if an effect that’s smaller than this size. A smaller effect size is unlikely to be detected in
# your study due to inadequate statistical power.
```

# What is a confidence interval? How to calculate the confidence interval, explain with an example.

```python
# A confidence interval is the mean of your estimate plus and minus the variation in that estimate. This is the range of values
# you expect your estimate to fall between if you redo your test, within a certain level of confidence. Confidence, in
# statistics, is another way to describe probability.

# Statisticians use confidence intervals to measure uncertainty in a sample variable. For example, a researcher selects 
# different samples randomly from the same population and computes a confidence interval for each sample to see how it may
# represent the true value of the population variable. The resulting datasets are all different where some intervals include
# the true population parameter and others do not.

# for example:
    
# Suppose you want to calculate the confidence interval for the average height of a sample of 50 individuals. You collect data
# and find that the sample mean height is 165 cm and the sample standard deviation is 10 cm.

# Step 1: Select the Confidence Level
# Let's use a 95% confidence level.

# Step 2: Collect and Analyze Data
# Sample Mean = 165 cm
# Sample Standard Deviation (s) = 10 cm
# Sample Size (n) = 50

# Step: 3
# Determine the Critical Value
# For a 95% confidence level and a normal distribution, the critical value is 1.96.
# Step 4: Calculate the Margin of Error (MOE)
    
# MOE(Margin Of Error)=1.96×(10/(50**0.5)) ≈2.78
# Step 5: Calculate the Confidence Interval
# Confidence Interval = 165 ± 2.78

# Confidence Interval=(162.22,167.78)


# Therefore, with 95% confidence, you can say that the true average height of the population is likely to fall between 
# 162.22 cm and 167.78 cm based on the sample data.
```

#  Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.


```python
# Bayes' Theorem is a mathematical formula that allows us to update the probability of an event based on new evidence or
# information. It's often used in probability and statistics to make predictions or decisions based on prior knowledge and 
# new data.

# The formula for Bayes' Theorem is as follows:

# P(A∣B)= P(B∣A)×P(A)/P(B)

# Where:

# P(A∣B) is the conditional probability of event A given event B.

# P(B∣A) is the conditional probability of event B given event A.

# P(A) is the prior probability of event A.

# P(B) is the prior probability of event B.

# Now, let's go through a sample problem and solution using Bayes' Theorem.

# Suppose you are a doctor and you want to determine the probability that a patient has a certain medical condition given the 
# results of a specific diagnostic test.

# The probability that a patient has the medical condition (P(A)) is 0.2.

# The probability that the diagnostic test is positive given that the patient has the medical condition (P(B∣A)) is 0.9.


# The probability that the diagnostic test is positive given that the patient does not have the medical condition 
# (P(B∣¬A)) is 0.1.

# want to calculate the probability that a patient has the medical condition (P(A∣B)) given that the diagnostic test is
#positive.

# P(A)=0.2

# P(B∣A)=0.9

# P(B∣¬A)=0.1

# P(B)=P(B∣A)×P(A)+P(B∣¬A)×P(¬A)

# P(B)=0.9×0.2+0.1×(1−0.2) = 0.18 + 0.08 = 0.26

# P(B)=0.18+0.08=0.26

# Now we can use Bayes' Theorem to calculate 

# P(A∣B)= P(B∣A)×P(A)/P(B)

# P(A∣B)= 0.9×0.2/0.26 = 0.692
```

# What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.


```python
# The margin of error (MOE) in a confidence interval (CI) represents the range around a point estimate (such as a sample mean 
# or proportion) within which the true population parameter is likely to fall with a certain level of confidence. In other 
# words, it quantifies the uncertainty or variability associated with the sample estimate.

# A larger margin of error indicates greater uncertainty, while a smaller margin of error indicates more precision in 
# estimating the population parameter.

# The formula to calculate the margin of error depends on the distribution being used (typically the normal distribution for
# large sample sizes) and the desired confidence level. For a given confidence level, the margin of error is determined by the
# critical value from the distribution and the standard error of the estimate.

# For example, in the case of a confidence interval for the population mean (μ) with a normal distribution and a large sample
# size (n), the margin of error (MOE) is calculated as:


# MOE=Critical Value × Standard Error

# Critical Value: Corresponds to the desired confidence level and the distribution being used. For a 95% confidence level in a
# normal distribution, the critical value is approximately 1.96.

# Standard Error: Represents the standard deviation of the sample estimate (such as the sample mean) divided by the square root
# of the sample size (n).

# In mathematical terms:

# MOE = Critical Value × Standard Deviation/(n)**0.5

# The margin of error gives you the interval around the sample estimate within which you are confident the true population
# parameter lies. It's a measure of the precision of your estimate.

# Margin of error is influenced by both the variability of the data (standard deviation) and the size of the
# sample. Larger sample sizes generally lead to smaller margins of error because the sample estimate becomes more accurate and
# less variable. Similarly, higher confidence levels result in wider intervals (larger margins of error) to account for 
# increased certainty.

# Interpreting the margin of error involves understanding that it provides a range within which the true population parameter 

# is likely to exist based on the sample data and the chosen level of confidence.


# Let's consider a scenario involving a political poll. Suppose a polling organization wants to estimate the proportion of 
# voters in a city who support a certain candidate in an upcoming mayoral election. The organization conducts two separate 
# polls with different sample sizes and compares the resulting margins of error.

# Scenario:
# Polling Organization A conducts a survey with a sample size of 300 voters.
# Polling Organization B conducts a separate survey with a sample size of 1000 voters.

# Both organizations aim to estimate the proportion of voters who support the candidate.

# Assumptions:
# The population of voters in the city is relatively large.
# The voters are randomly selected.
# The true proportion of voters who support the candidate is unknown.

# Calculating the Margins of Error:
# For simplicity, let's assume that both polls yield a sample proportion of 0.55 (55% support) with a standard deviation of
#  0.5.This is just an example, and in real scenarios, the sample proportion and standard deviation may vary.

# The formula for the margin of error for a proportion is:

# MOE= Critical Value × (p(1−p)/n)**0.5

# Where:
# p is the sample proportion.
# n is the sample size.
# The critical value depends on the desired confidence level and is approximately 1.96 for a 95% confidence level.

# Calculations:
# For both organizations, let's calculate the margin of error for a 95% confidence level using the given assumptions:

# For Polling Organization A (Sample Size = 300, Sample Proportion = 0.55):

# MOE(A) = 1.96×(0.55×(1−0.55)/300)**0.5 ≈0.068

# For Polling Organization B (Sample Size = 1000, Sample Proportion = 0.55):

# MOE(B) = 1.96×(0.55×(1−0.55)/1000)**0.5 ≈0.044

# In this scenario, Polling Organization A has a smaller sample size (300) compared to Polling Organization B (1000). As a 
# result, Polling Organization B has a smaller margin of error (0.044) compared to Polling Organization A (0.068).

# This means that Polling Organization B's estimate of the proportion of voters who support the candidate is likely to be more
# precise and closer to the true proportion in the population. The larger sample size allows for a more accurate estimation, 
# resulting in a smaller margin of error and greater confidence in the estimate.
```

# In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.



```python
# given 
# sample size(n) = 50
# sample mean = 6 
# sample std = 2.5
# C.I = 95 %
# In this question population mean is not given. So if we assume population mean(weight loss) = 0

# Step 1:
# Null Hypothesis(H0): Drug is not effective, population mean(weight loss) = 0 
# Alternate Hypothesis:Drug is Effective, population mean(weight loss) != 0( 2 Tail Test)  

# Step 2:
# alpha = 0.05

# Step 3:
# degree of freedom= n-1= 50- 1=49

# Step 4:
# at 95% C.I and dof =49 , if t test is less than -2.0096 and greater than 2.0096  reject the Null Hypothesis

# Step 5:
    
# t= (sample mean - population mean)/std/(n)**0.5
# t= (6 - 0)/2.5/(50)**0.5 =16.97
# So , We can reject the Null Hypothesis So, Drug is Significantly Effective.
```
